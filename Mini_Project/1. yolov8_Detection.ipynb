```python
import os
from ultralytics import YOLO
```

```python
def check_txt_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            if os.path.getsize(file_path) == 0:
                print(f"파일 '{filename}'는 비어 있습니다.")
            else:
                print(f"파일 '{filename}'는 비어 있지 않습니다.")

# 검사할 디렉토리 경로를 지정합니다.
directory_path = "C:/Users/lim1007/Desktop/yoloo/data/hohoi_label"

check_txt_files(directory_path)
```

```python
txt_directory_path = "C:/Users/lim1007/Desktop/yoloo/data/hohoi_label/"
image_directory_path = "C:/Users/lim1007/Desktop/yoloo/data/hohoi_img/"

def check_matching_files(txt_directory, image_directory):
    txt_files = []
    jpg_files = []

    for filename in os.listdir(txt_directory):
        if filename.endswith(".txt"):
            txt_files.append(filename)

    for filename in os.listdir(image_directory):
        if filename.endswith(".jpg"):
            jpg_files.append(filename)

    for txt_file in txt_files:
        if txt_file[:-4] not in [jpg_file[:-4] for jpg_file in jpg_files]:
            print(f"텍스트 파일 '{txt_file}'와 매칭되는 이미지 파일이 없습니다.")

    for jpg_file in jpg_files:
        if jpg_file[:-4] not in [txt_file[:-4] for txt_file in txt_files]:
            print(f"이미지 파일 '{jpg_file}'와 매칭되는 텍스트 파일이 없습니다.")

check_matching_files(txt_directory_path, image_directory_path)
```

```python
# YAML 파일 경로를 정의합니다.
yaml_path = os.path.join("C:/Users/lim1007/Desktop/yoloo/yayaml.yaml")

# 설정 파일로 YOLO 객체를 초기화.
model = YOLO("yolov8n.yaml")
#  patience=30,
# 데이터와 지정한 에폭으로 YOLO 모델을 학습시킵니다.
train_results = model.train(data=yaml_path, epochs=5, patience=30, batch=8, imgsz=416)
```
